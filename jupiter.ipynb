```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

df=pd.DataFrame({
    
    'Name':["Amruth","Aman","Kavin","Tushar","Rohith"],
    'Marks':[56,87,75,80,95]
    
})
print(df)
```

```python
x=df['Name']
y=df['Marks']
plt.plot(x,y,color="red")
plt.title("Student Marks Detial")
plt.xlabel("Students")
plt.ylabel("Marks")
```
